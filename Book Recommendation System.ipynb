## Book Recommendation system using Clustering and Collabrative based filtering Technique

```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv(
    "C:\\Users\\HP-PC\\OneDrive\\Documents\\Data Science\\Book Recommendation dataset\\BX-Books.csv",
    sep=";",
    encoding="latin-1",
    on_bad_lines="skip"  # Skips problematic rows
)
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-L']]
```

```python
books.rename(columns = {'Book-Title': 'Title', 'Book-Author' : 'Author',  'Year-Of-Publication': 'Year', 'Image-URL-L': 'Image'}, inplace = True )
```

```python
users = pd.read_csv(r"C:\Users\HP-PC\OneDrive\Documents\Data Science\Book Recommendation dataset\BX-Users.csv", sep=";",
    encoding="latin-1",
    on_bad_lines="skip"  # Skips problematic rows
)
```

```python
ratings = pd.read_csv(r"C:\Users\HP-PC\OneDrive\Documents\Data Science\Book Recommendation dataset\BX-Book-Ratings.csv", sep=";",
    encoding="latin-1",
    on_bad_lines="skip"  # Skips problematic rows
    )
```

```python
ratings
```

```python
ratings.rename(columns = {'User-ID':'user_id', 'Book-Rating' :'book_rating'}, inplace = True)
```

```python
x = ratings['user_id'].value_counts()>200
```

```python
x
```

```python
y= x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings
```

```python
rwb = ratings.merge(books, on = "ISBN")
```

```python
num_rating = rwb.groupby('Title')['book_rating'].count().reset_index()
```

```python
num_rating = num_rating.rename(columns = {'book_rating':'num_of_rating'})
```

```python
num_rating
```

```python
final_rating = rwb.merge(num_rating, on = 'Title')
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.drop_duplicates(['user_id', 'Title'], inplace = True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns = 'user_id', index = 'Title', values = 'book_rating')
```

```python
final_rating.columns
```

```python
book_pivot 
```

```python
book_pivot.fillna(0, inplace = True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model =  NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors = 6)
```

```python
distance
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
book_name = book_pivot.index
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors = 6)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
    
```

```python
book_name = 'A Bend in the Road'
recommend_book(book_name)
```
